```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("general_data.csv")
df.head()
```

```python
df.corr()
```

```python
#droping few unneccesary features 
df.drop(['BusinessTravel', 'EducationField', 'EmployeeID', 'JobRole', 'EmployeeCount', 'Over18', 'StandardHours'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.fillna('NumCompaniesWorked' == 0 )  # Filled null values
```

```python
df.describe().T  #Describing the data set
```

```python
# Converting the text variables to the int datatypes
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Attrition'] = label_encoder.fit_transform(df['Attrition'])  #Converted Attrition to int type from object type
df['Department'] = label_encoder.fit_transform(df['Department']) #converted department to int type as(1,2,3)
df['Gender'] = label_encoder.fit_transform(df['Gender']) #converted gender to int type 
df['MaritalStatus'] = label_encoder.fit_transform(df['MaritalStatus']) #converted marital status to int type
df['NumCompaniesWorked'] = label_encoder.fit_transform(df['NumCompaniesWorked'])
df['TotalWorkingYears'] = label_encoder.fit_transform(df['TotalWorkingYears'])
```

```python
df.head()
```

# Choosing the IDV And DV

```python

y  = df['Attrition']  #dependent variable
x = df[['Age', 'Department', 'DistanceFromHome', 'Education', 'Gender', 'JobLevel', 'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']] #independent variable
```

```python
import statsmodels.api as sm
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
Result = Logistic.fit()
```

```python
Result.summary()
```

# Random Forest

```python
from sklearn import preprocessing
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df.columns
```

```python
rf_model = RandomForestClassifier(n_estimators = 1000 , max_features = 2 , oob_score = True)
```

```python
features = ['Age', 'Department', 'DistanceFromHome', 'Education', 'Gender', 'JobLevel', 'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
rf_model.fit(X = df[features] , y = df["Attrition"])
```

```python
RandomForestClassifier
```

```python
print(rf_model.oob_score_);
```

```python
importances = rf_model.feature_importances_
```

```python
for feature , imp in zip(features , rf_model.feature_importances_):
    print(feature,imp)
```
