### wilcoxon test

```python
from scipy.stats import wilcoxon
```

```python
data.head()
```

```python
d1 = data.Attrition
d2 = data.BusinessTravel
```

```python
stat, p = wilcoxon(d1,d2)
```

```python
print(stat , p)
```

```python
# taking another data comparison  # 2nd example
```

```python
d1 = data.DistanceFromHome
d2 = data.PercentSalaryHike
```

```python
stat, p = wilcoxon(d1,d2)
```

```python
print(stat , p)
```

```python
# 3 rd example
```

```python
d1 = data.TotalWorkingYears
d2 = data.Department
```

```python
stat, p = wilcoxon(d1,d2)
```

```python
print(stat , p)
```

```python
print(f'The Value of p in wilcoxon test is: {p}')
if p < 0.05:
    print('There is significant impact of TotalWorkingYears on Department')
else:
    print('There is no significant impact of TotalWorkingYears on Department')
```
