```python
# lets use python knowledge here
# use for loop to execute the remaining variables and to compare the variables
```

```python
for i in data.columns:
    if i not in ['Attrition','EmployeeID','EmployeeCount','StandardHours','NumCompaniesWorked','TotalWorkingYears']: 
        temp , p = pearsonr(data.Attrition , data[i])
        print(i , end = ": ")
        print(temp , p , end = "\n")
        if(p<0.05):
            print("reject null hypothesis \n")
        else:
            print(" accept null hypothesis\n")
```

```python
# to check
# lets find and compare the variable 
stat3,p3=pearsonr(data.Attrition,data.YearsAtCompany)
print(stat3,p3)
if(p< 0.05):
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
